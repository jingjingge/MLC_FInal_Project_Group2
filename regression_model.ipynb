# model fine tune
Hyperparameter Tuning for selected model

#### decision tree

```python
%%time
# Creating a dictionary of parameters to use in GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

dt_params = {
    'criterion':  ['gini', 'entropy'],
    'max_depth':  [None, 2, 4, 6, 8, 10],
    'max_features': [None, 'sqrt', 'log2', 0.2, 0.4, 0.6, 0.8],
    'splitter': ['best', 'random']
}

dt = RandomizedSearchCV(
                   estimator=DecisionTreeClassifier(),
                   param_distributions=dt_params,
                   n_iter = 200,
#                    scoring='balanced_accuracy',
                   scoring='accuracy',    
                   cv = 5,
                   n_jobs = -1,
                   verbose = 2,
                   random_state = 2022)

dt.fit(X_train, y_train)
print('Best accuracy: %.3f' % dt.best_score_)
print('\nBest params:\n', dt.best_params_)
```

#### random forest

```python
%%time
# Create the grid parameter:
rf_params = {'n_estimators': [100, 120, 150],
           'criterion': ['entropy', 'gini'], 
           'max_depth': [None,1,3,5,7,9],
           'max_features': range(1,11),  
           'min_samples_split': range(2, 10),
           'min_samples_leaf': [1,3,5]}

# Create the grid:
rf = RandomizedSearchCV(
                  estimator=RandomForestClassifier(),
                  param_distributions= rf_params,
                  n_iter = 200, 
#                   scoring='balanced_accuracy',
                  scoring='accuracy',
                  cv=5, 
                  n_jobs=-1,
                  verbose =2,
                  random_state = 2022)

# Fit using grid search:
rf.fit(X_train, y_train)

# Print best accuracy and best parameters:
print('Best accuracy: %.3f' % rf.best_score_)
print('\nBest params:\n', rf.best_params_)
```

#### xgboost

```python
%%time
import xgboost as xgb
# Create the grid parameter:
xgb_params = {'max_depth': [3, 6, 10, 15],
              'learning_rate': [0.01, 0.1, 0.2, 0.3, 0.4], 
              'subsample': np.arange(0.5, 1.0, 0.1),
              'colsample_bytree': np.arange(0.5, 1.0, 0.1),
              'colsample_bylevel': np.arange(0.5, 1.0, 0.1),
              'n_estimators': [100, 250, 500, 750],
              'num_class': [10],
              'tree_method':['auto', 'exact', 'approx', 'hist']}

xgboost = RandomizedSearchCV(
                        estimator=xgb.XGBClassifier(objective='multi:softmax'),
                        param_distributions = xgb_params,
                        n_iter = 200,
                        scoring='accuracy',
#                         scoring='balanced_accuracy',
                        cv=5,
                        n_jobs=-1,
                        verbose=2)

xgboost.fit(X_train, y_train)

# Print best accuracy and best parameters:
print('Best accuracy: %.3f' % xgboost.best_score_)
print('\nBest params:\n', xgboost.best_params_)
```
