# CLASSIFICATION


```python
Y_name = 'urination_density_pop_degree'
```

```python
y = df2[Y_name]
X = df2.drop(['urination_count','urination_density_pop' ,'Zipcode','urination_density_area','urination_density_pop_degree'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

SEED = 123
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=SEED)
```

# model fine tune
Hyperparameter Tuning for selected model

#### decision tree

```python
%%time
# balanced_accuracy
# Creating a dictionary of parameters to use in GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

dt_params = {
    'criterion':  ['gini', 'entropy'],
    'max_depth':  [None, 2, 4, 6, 8, 10],
    'max_features': [None, 'sqrt', 'log2', 0.2, 0.4, 0.6, 0.8],
    'splitter': ['best', 'random']
}

dt = RandomizedSearchCV(
                   estimator=DecisionTreeClassifier(),
                   param_distributions=dt_params,
                   n_iter = 200,
                   scoring='balanced_accuracy',
                   cv = 5,
                   n_jobs = -1,
                   verbose = 2,
                   random_state = 2022)

dt.fit(X_train, y_train)
print('Best accuracy: %.3f' % dt.best_score_)
print('\nBest params:\n', dt.best_params_)
```

#### xgboost

```python
%%time
# accuracy
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb
# Create the grid parameter:
xgb_params = {'max_depth': [3, 6, 10, 15],
              'learning_rate': [0.01, 0.1, 0.2, 0.3, 0.4], 
              'subsample': np.arange(0.5, 1.0, 0.1),
              'colsample_bytree': np.arange(0.5, 1.0, 0.1),
              'colsample_bylevel': np.arange(0.5, 1.0, 0.1),
              'n_estimators': [100, 250, 500, 750],
              'tree_method':['auto', 'exact', 'approx', 'hist']}

xgboost = RandomizedSearchCV(
                        estimator=xgb.XGBClassifier(objective='multi:softmax',num_class = 3),
                        param_distributions = xgb_params,
                        n_iter = 200,
                        scoring='accuracy',
                        cv=5,
                        n_jobs=-1,
                        verbose=2)

xgboost.fit(X_train, y_train)

# Print best accuracy and best parameters:
print('Best accuracy: %.3f' % xgboost.best_score_)
print('\nBest params:\n', xgboost.best_params_)
```

# Final model evaluation

DecisionTreeClassifier

```python
%%time
# DecisionTreeClassifier
# 选balanced_accuracy作为标准
# Create the final Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt_final = DecisionTreeClassifier(splitter = 'random',
                                  max_features = 0.4,
                                  max_depth = 10,
                                  criterion = 'entropy')

# Train the models using 80% of the original data
dt_final.fit(X_train, y_train)

y_pred = dt_final.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d').set_title('urination_count confusion matrix (0 = low risk, 1 = medium risk, 2 = high risk)')

print(classification_report(y_test,y_pred))
```
